### Bài 2b: Tìm số nhỏ nhất của 1 list và vị trí của nó, nếu có nhiều số nhỏ nhất bằng nhau thì lấy chỉ số đầu tiên (nhỏ nhất)
10 0 0 12 34 5 6 7  8 89 
0 1

```python
def min_and_pos(lst):
    if lst:
        min_value=lst[0]
        min_pos=1
        for i,x in enumerate(lst,1):
            if x<min_value:
                min_value=x
                min_pos=i
        return min_value,min_pos
    return None
assert min_and_pos([10,0,0,12,34,5,6,7,8,89,0,1])==(0,2)
assert min_and_pos([])==None
print("Pass all Tests!")
```

### Bài 3: Tìm giá trị trung bình của các số nguyên dương trong 1 list
Nếu không có số dương nào thì trả về "invalid"

```python
def trung_binh(lst):
    if lst:
        t=0
        d=0
        for x in lst:
            if x>0:
               t+=x
               d+=1
        if not d:
            return "invalid"
        return t/d
    return "invalid"
lst=list(map(int,input().split()))
print(trung_binh(lst))
```

### Bài 4: Kiểm tra một list số nguyên có phải là đối xứng hay không

```python
lst1=list(map(int,input().split()))
def doi_xung(lst):
    if lst:

        l=len(lst)
        for i in range(l//2):
            if lst[i] != lst[l-1-i]:
                return False
        return True
    return None
print(doi_xung(lst1))
```

### Bài 5: Đưa tất cả các số bằng 0 trong list về đầu của list. Các số khác không thay đổi thứ tự trước sau
3 4 6 7 1 9 0 0 23 0 0 34 67 89
0 0 0 0 3 4 6 7 1 9 23 34 67 89

```python
lst=list(map(int,input().split()))
def dua_so_0(lst):
    zero=[]
    others=[]
    for x in lst:
        if x==0:
            zero.append(x)
        else:
            others.append(x)
    new_lst=zero+others
    return new_lst
print(*dua_so_0(lst))

```

### Bài 6: Xóa tất cả các số giống nhau trong list, Các số khác giữ nguyên vị trí trước sau

```python
lst=list(map(int,input().split()))
def so_giong_nhau(lst):
    if lst:
        ans=[]
        for x in lst:
            if x not in ans:
                ans.append(x)
        return ans
    return None
print(so_giong_nhau(lst))
```

### Bài 8: Kiểm tra một list số nguyên có phải là ziczac khong
1 3 2 5 2 8 7 9 2 100: ziczac
34 2 56 7 90 20 30 7 120: ziczac

```python
lst=list(map(int,input().split()))
def ziczac(lst):
    if lst:
        for i in range(1,len(lst)-1):
            if (lst[i]-lst[i-1])*(lst[i]-lst[i+1])<=0:
                return False
        return True
print(ziczac(lst))
```

### Bài 9: Tìm và in ra tất cả các vị trí của số nguyên k trong một danh sách số nguyên cho trước. Nếu không có k trong ds thì in ra "-1".
VD: k = 3, [1,3,4,4,3,3,4,3,6,1] => k xuất hiện tại các vị trí (trả về 1 list) [1, 4, 5, 7]
k = 10, [1,3,4,4,3,3,4,3,6,1] => k không xuất hiện trong ds => trả về "-1"

```python
lst=list(map(int,input().split()))
k=int(input())
vi_tri=[]
for i in range(len(lst)):
    if lst[i] ==k:
        vi_tri.append(i)
if len(vi_tri)==0:
    print("-1")
else:
    print(vi_tri)
```
