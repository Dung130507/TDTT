Bài 3: Viết chương trình nhập vào một “kí tự” c, nếu nó là chữ hoa thì in ra chữ thường 
tương ứng và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó. (xem 
trước cấu trúc lựa chọn: if….else, ma) 

```python
c=input()
def ki_tu(c):
    if len(c)==1 and ord("A")<=ord(c)<=ord("Z"):
        return chr(ord(c)+32)
    elif len(c)==1 and ord("a")<=ord(c)<=ord("z"):
        return chr(ord(c)-32)
    else:
        return ("Nhap ki tu")
print (ki_tu(c))
```

Bài 4: Viết chương trình nhập vào một “kí tự”, Kiểm tra xem c có phải là kí tự chữ cái 
không (hoa hoặc thường). Nếu đúng thì in ra “{c} là kí tu alphabet”, nếu sai in ra “{c} 
không phải là kí tự alphabet.

```python
c=input()
def ki_tu(c):
    if len(c)==1 and ord("A")<=ord(c)<=ord("Z") or ord("a")<=ord(c)<=ord("z"):
        return f"{c} la ki tu alphabet"
    else:
        return f"{c} khong phai ki tu alphabet"
print (ki_tu(c))
```

Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không, 
nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó 
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”. 

```python
a=int(input())
b=int(input())
c=int(input())
def tam_giac(a,b,c):
    if a+b>c and a+c>b and b+c>a :
        d = (a+b+c)/2
        import math
        e = math.sqrt(d*(d-a)*(d-b)*(d-c))
        return round(e, 1)
    else:
        return ("Khong phai 3 canh cua tam giac")
print (tam_giac(a,b,c))
```

Bài 7: Nhập vào một chuỗi gồm các kí tự thường độ dài >=20, in ra chữ cái thứ năm và 
thứ chín của chuỗi này

```python
chuoi_ban_dau=input()
gio_chu_cai=[]
if len(chuoi_ban_dau)<20:
    print ("Chuỗi nhập vào quá ngắn")
else:
    for ky_tu in chuoi_ban_dau:
        if ky_tu.isalpha():
            gio_chu_cai.append(ky_tu)
    so_luong=len(gio_chu_cai)
    if so_luong>=9:
        chu_thu_5=gio_chu_cai[4]
        chu_thu_9=gio_chu_cai[8]
        print(f"Tổng số chữ cái lọc được:{so_luong}")
        print(f"chữ cái thứ 5 là {chu_thu_5} ")
        print(f"chữ cái thứ 9 là {chu_thu_9} ")
    else:
        print("Khong du 9 chu cai")
```
