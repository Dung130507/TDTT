W3A1  Viết một chương trình nhập vào một số nguyên n và in ra các chữ số đảo ngược nhau

```python
n=input()
if n.isdigit():
    print (int(n[::-1]))
elif n[0] == "-" and n[1].isdigit():
    print ("-" + n[1:][::-1])
else:
    print ("Nhap 1 so nguyen")
```

[Longer string] 
Nhập vào hai chuỗi a, b gồm các chữ cái thường. In ra "True" nếu chuỗi a dài hơn chuỗi b. Ngược lại in ra "False".

```python
a=input()
b=input()
chu_cai_a=[]
chu_cai_b=[]
for ky_tu in a:
    if ky_tu.isalpha():
        chu_cai_a.append(ky_tu)
for kt in b:
    if kt.isalpha():
        chu_cai_b.append(kt)
print (len(chu_cai_a)>len(chu_cai_b))
```
