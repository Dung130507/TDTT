W4A6 
"[SumPrimesInRange] 
Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn 
[a, b]. Mỗi dòng là một bộ test (a b)."

```python
import math
def snt(n):
    if n<2:
        return False
    for i in range(2,int(math.sqrt(n))+1):
        if n%i==0:
            return False
    return True
a,b=map(int,input().split())
danh_sach_snt=[x for x in range (a,b+1) if snt(x)]
tong=sum(danh_sach_snt)
print (f"Cac so nguyen to trong doan [a,b] la {danh_sach_snt}")
print(f"Tong cac so nguyen to trong doan [a,b] la {tong}")
```

W4A7 
"[LargestPrimeFactor] 
Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n." 


```python
n=int(input())
if n<2:
    print("Nhap lai")
else:
    so=n  #Tạo biến tạm "so" để thực hiện phép chia dần (bào mòn)
    uoc_snt=[]   #Khởi tạo danh sách rỗng để lưu các snt tìm được
    i=2         #Bắt đầu kiểm tra từ snt nhỏ nhất là 2
    #Bước 3:Chỉ chạy đến bậc 2 của "so"
    #Nếu i*i>so thì phần còn lại của "so" chắc chắn là snt
    while i*i<=so:
        #Trong khi "so" vẫn còn chia hết cho i
        while so%i==0:
            #Nếu i chưa có trong danh sách thì mới thêm vào (tránh trùng lặp)
            if i not in uoc_snt:
                uoc_snt.append(i)
            #Phép chia nguyên để rút gọn "so" sau khi đã tìm thấy ước i
            so //= i
        #Tăng i lên để kiểm tra số tiếp theo
        i +=1
    #Bước 4:Xử lý trường hợp "so" còn lại sau vòng lặp
    # Nếu so > 1 thì giá trị cuối cùng này chính là ước số nguyên tố cuối cùng   
    if so>1:
        uoc_snt.append(so)
    print (f"uoc so nguyen to lon nhat cua {n} la {max(uoc_snt)}")
```

W4A8 
ReverseAddPalindrome] 
Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số 
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được.

```python
limit=1000
buoc=0
n_goc=int(input())
n=n_goc
while True:
    dao=int(str(n)[::-1])
    if n==dao or buoc>=limit:
        break
    n=n+dao
    buoc += 1
if buoc<limit:
    print (f"so doi xung tu {n_goc} la {n} va so buoc la {buoc} ")
else:
    print ("Khong tim thay")
```

W4A10 
"[LongestCollatzUpToN] 
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất 
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số 
nhỏ nhất."

```python
n=int(input())
max_L=0
best_x=0
for i in range(1, n + 1):
    curr_n = i
    curr_L = 1 
    while curr_n != 1:
        if curr_n % 2 == 0:
            curr_n //= 2
        else:
            curr_n = 3 * curr_n + 1
        curr_L += 1
    if curr_L > max_L:
        max_L = curr_L
        best_x = i
print (f"{best_x} {max_L}")
```

[ClosePairNumber] 
Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật: 
Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại. 
Ví dụ 220 và 284 là một cặp số thân thiết 
220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55, 
110. Tổng của 11 ước này cho chúng ta 284 
284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng 
vừa đúng bằng 220 
Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không. 
Nếu có, in ra màn hình "true", ngược lại, in ra "false"

```python
import math
def tinh_tong_uoc(n):
    tong=1
    for i in range (2,int(math.sqrt(n))+1):
        if n%i==0:
            tong +=i
            j=n//i
            if j!=i:
                tong +=j
    return tong
a,b=map(int,input().split())
s1=tinh_tong_uoc(a)
s2=tinh_tong_uoc(b)
print(s2==a and s1==b)
```
