### Problem 4
#### Write a function that returns the count of positive integers k such that 1 <= k < n and k > (n - 1) / 2. (Example: n=9 -> 4).

```python
n =int(input())
def dem_so_nguyen(n):
    count = 0 
    for k in range(1, n): 
        if k > (n - 1) / 2: 
            count += 1  
    return count

print(dem_so_nguyen(n))
```

### Problem 8
#### Write a function that returns the character that is n positions after character x in the alphabet.
#### Preserve case and wrap around within A..Z or a..z.

```python
x=input()
n=int(input())
vt=ord(x)+n
def chu_cai(x,n):
    if ord("A")<=vt<=ord("Z"):
        return chr(vt)
    elif ord("a")<=vt<=ord("z"):
        return chr(vt)
    else:
        return ("Khong phai chu cai")
print (chu_cai(x,n))
    
```

### Problem 9
#### Write a function that converts an uppercase letter to lowercase. If the input is not a single character, raise ValueError.

```python
x=input()
def chu_cai(x):
    if len(x)==1 and ord("A")<=ord(x)<=ord("Z"):
        return chr(ord(x)+32)
    else:
        raise ValueError
print (chu_cai(x))
```

### Problem 12
#### Write a function that returns a right isosceles triangle of height n using '*'.
#### Line i (1-indexed) has i '*', joined with newlines.
#### Example with n = 4
#### *
#### **
#### ***
#### ****

```python
n=int(input())
def tam_giac_vuong_can(n):
    ket_qua=("")
    for i in range(1,n+1):
        dong_hien_tai="*"*i
        ket_qua+= dong_hien_tai + "\n"
    return ket_qua
print (tam_giac_vuong_can(n))
```
